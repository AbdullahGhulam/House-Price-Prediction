# House Price Prediction using Linear Regression

## Project Overview
This project demonstrates a complete machine learning workflow for predicting house prices based on key demographic and geographic features using linear regression. The model utilizes multiple features including area income, house age, number of rooms, number of bedrooms, and population density to predict residential property prices.

## Methodology
1. **Data Loading & Exploration**: Load dataset and examine statistical properties
2. **Exploratory Data Analysis (EDA)**: Understand feature relationships and data distribution
3. **Feature Selection**: Identify and select relevant features for the model
4. **Model Building**: Train a linear regression model using scikit-learn
5. **Evaluation**: Assess model performance using multiple regression metrics
6. **Visualization**: Analyze model predictions and residuals

```python
# ============================================================================
# IMPORTS
# ============================================================================
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score, mean_absolute_error

# Configure visualization settings
plt.style.use('seaborn-v0_8-darkgrid')
sns.set_palette('husl')
%matplotlib inline
```

## 1. Data Loading & Exploration

```python
# Load the housing dataset
df = pd.read_csv('USA_Housing.csv')

# Display first few rows
print("Dataset Overview:")
display(df.head())

# Display basic statistics
print("\nDataset Shape:")
print(f"Rows: {df.shape[0]}, Columns: {df.shape[1]}")
```

```python
# Examine data types and missing values
print("Dataset Information:")
print(df.info())
print("\nMissing Values:")
print(df.isnull().sum())

# Remove the 'Address' column as it is not useful for numerical modeling
df.drop('Address', axis=1, inplace=True)
print("\n'Address' column removed from dataset.")
```

```python
# Display descriptive statistics
print("Descriptive Statistics:")
display(df.describe())
```

## 2. Exploratory Data Analysis (EDA)

```python
# Analyze feature correlations
print("Feature Correlation with Target:")
correlation_with_price = df.corr()['Price'].sort_values(ascending=False)
print(correlation_with_price)

# Visualize correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(df.corr(), annot=True, cmap='coolwarm', fmt='.2f', square=True)
plt.title('Feature Correlation Matrix', fontsize=14, fontweight='bold')
plt.tight_layout()
plt.show()
```

## 3. Model Development

```python
# Define features and target variable
feature_columns = [
    'Avg. Area Income',
    'Avg. Area House Age',
    'Avg. Area Number of Rooms',
    'Avg. Area Number of Bedrooms',
    'Area Population'
]

X = df[feature_columns]
y = df['Price']

print(f"Features shape: {X.shape}")
print(f"Target shape: {y.shape}")
```

```python
# Split data into training and testing sets (70-30 split)
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.3, random_state=101
)

print(f"Training set size: {X_train.shape[0]}")
print(f"Testing set size: {X_test.shape[0]}")
```

```python
# Initialize and train the Linear Regression model
model = LinearRegression()
model.fit(X_train, y_train)

# Make predictions on test set
y_pred = model.predict(X_test)

print("Model trained successfully.")
```

## 4. Model Evaluation

```python
# Calculate evaluation metrics
mse = mean_squared_error(y_test, y_pred)
rmse = np.sqrt(mse)
mae = mean_absolute_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

# Display evaluation results
print("\n" + "="*50)
print("MODEL PERFORMANCE METRICS")
print("="*50)
print(f"R² Score:                  {r2:.4f}")
print(f"Mean Absolute Error (MAE):  ${mae:,.2f}")
print(f"Mean Squared Error (MSE):   ${mse:,.2f}")
print(f"Root Mean Squared Error:    ${rmse:,.2f}")
print("="*50)

# Display feature coefficients
print("\nModel Coefficients:")
for feature, coef in zip(feature_columns, model.coef_):
    print(f"{feature:.<40} {coef:>12,.2f}")
print(f"{'Intercept':.<40} {model.intercept_:>12,.2f}")
```

## 5. Model Visualization & Analysis

```python
# Residual Plot: Actual vs Predicted Values
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Scatter plot of actual vs predicted
axes[0].scatter(y_test, y_pred, alpha=0.6, s=50, edgecolors='k')
axes[0].plot([y_test.min(), y_test.max()], [y_test.min(), y_test.max()], 
             'r--', lw=2, label='Perfect Prediction')
axes[0].set_xlabel('Actual Price ($)', fontsize=11)
axes[0].set_ylabel('Predicted Price ($)', fontsize=11)
axes[0].set_title('Predicted vs Actual House Prices', fontsize=12, fontweight='bold')
axes[0].legend()
axes[0].grid(True, alpha=0.3)

# Residual distribution plot
residuals = y_test - y_pred
axes[1].scatter(y_pred, residuals, alpha=0.6, s=50, edgecolors='k')
axes[1].axhline(y=0, color='r', linestyle='--', lw=2)
axes[1].set_xlabel('Predicted Price ($)', fontsize=11)
axes[1].set_ylabel('Residuals ($)', fontsize=11)
axes[1].set_title('Residual Plot', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

```python
# Residual distribution
fig, axes = plt.subplots(1, 2, figsize=(14, 5))

# Histogram of residuals
axes[0].hist(residuals, bins=30, edgecolor='black', alpha=0.7, color='steelblue')
axes[0].set_xlabel('Residual Value ($)', fontsize=11)
axes[0].set_ylabel('Frequency', fontsize=11)
axes[0].set_title('Distribution of Residuals', fontsize=12, fontweight='bold')
axes[0].grid(True, alpha=0.3, axis='y')

# Q-Q plot for normality check
from scipy import stats
stats.probplot(residuals, dist="norm", plot=axes[1])
axes[1].set_title('Q-Q Plot: Residual Normality', fontsize=12, fontweight='bold')
axes[1].grid(True, alpha=0.3)

plt.tight_layout()
plt.show()
```

## 6. Conclusions

The linear regression model demonstrates strong predictive performance with an R² score of {:.4f}, indicating that the selected features explain approximately {:.1f}% of the variance in house prices. The model's coefficients show that:

- **Area Income** has the strongest positive relationship with price
- **House Age** and **Area Population** also contribute significantly to price predictions
- **Mean Absolute Error** of ${:,.0f} suggests the model's predictions are typically off by this amount

The residual analysis indicates relatively symmetric distribution with most residuals concentrated around zero, suggesting the linear assumption is reasonably valid for this dataset.
